```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label encoding of breast cancer db
loading data from the csv file

```python
cancer_data = pd.read_csv('data.csv')
```

```python
cancer_data.head()
```

```python
#finding the count of the different labels
cancer_data['diagnosis'].value_counts()
```

```python
#load the label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#appending the labels to the dataframs
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

0 ---> Benign
1 ---> Malignant


```python
cancer_data['target'].value_counts()
```

```python
iris_data= pd.read_csv('iris_data.csv')
```

```python
iris_data.head()
```

```python
iris_data['Species'].value_counts()
```

```python
#loading the label encoder
label_encoder_iris = LabelEncoder()
```

```python
iris_labels = label_encoder_iris.fit_transform(iris_data.Species)
```

```python
iris_data['target'] = iris_labels
```

```python
iris_data.head()
```

```python
# @title Species

from matplotlib import pyplot as plt
import seaborn as sns
iris_data.groupby('Species').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title SepalWidthCm vs PetalLengthCm

from matplotlib import pyplot as plt
iris_data.plot(kind='scatter', x='SepalWidthCm', y='PetalLengthCm', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```
